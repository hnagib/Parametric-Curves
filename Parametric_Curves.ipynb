# B-Spline Curves

$$P(u) =\sum\limits_{i=0}^n P_i N_{i,k}(u) \qquad t_{k-1} \leq u \leq t_{n+1}$$

This is the equation for the curve given by $n+1$ control points. $k$ is the order of the curve. The degree of the curve is $k-1$.  

Non-periodic uniform knot values are given by:

$$
t_i=
\begin{cases}
0 \qquad \qquad \quad 0 \leq i < k\\
i-k+1 \qquad k \leq i \leq n\\
n-k+2 \qquad n < i \leq n+k
\end{cases}
$$

Blending functions are produced recursively from the knot values:

$$
N_{i,1}(u)=
\begin{cases}
1 \qquad t_i \leq u \leq t_{i+1}\\
0 \qquad otherwise
\end{cases}
$$

$$N_{i,k}(u) = \frac{u-t_i}{t_{i+k-1}-t_i}N_{i,k-1}(u) + \frac{t_{i+k}-u}{t_{i+k}-t_{i+1}}N_{i+1,k-1}(u) $$

The code below automatically generates non-priodic uniform knot values for a given set of control points and plots the resulting B-Spline curve and it's blending functions of different degrees.

```python
pylab.rcParams['figure.figsize'] = (12.0, 12.0)

# List of control points to use
P_i = np.array([[-2,-1,3,4,1,0,2,0],        # x-coordinates of control points
                [0,2,2,0,1,0.5,-0.5,-0.5]]) # y-coordinates of control points

B_spline_interact(ctrl_points = P_i,        # List all control points
                  pt  = 6,                  # Number of control points to use
                  k   = 3,                  # Order of the curve (degree = k-1)
                  t   = 0.5,                # Let 0 < u < 1, Plot a point at u = t 
                  P1x = -1,                 # x-value of P_1
                  P1y = 2)                  # y-value of P_1

# Note: Uncomment the line below to use ipywidgets interactive features
# interact(B_spline_interact, ctrl_points = 0, pt = (3,8,1), k  = (2,8,1), t  = (0,0.99,.01), P1x = (-5,5,.01), P1y = (-5,5,.01));
```
